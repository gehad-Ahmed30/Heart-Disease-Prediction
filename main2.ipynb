```python
q1=data['SleepHours'].quantile(0.25)
q3=data['SleepHours'].quantile(0.75)
IQR=q3-q1
lower_bound=q1-1.5*IQR
upper_bound=q3+1.5*IQR
data = data[(data['SleepHours'] >= lower_bound) & (data['SleepHours'] <= upper_bound)]
data
```
